```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.utils import to_categorical, plot_model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import BatchNormalization, Conv2D, MaxPooling2D, Activation, Flatten, Dropout, Dense, Input
from tensorflow.keras import backend as K
from tensorflow.keras.callbacks import ReduceLROnPlateau
from sklearn.utils.class_weight import compute_class_weight
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import random
import cv2
import os
import glob

# initial parameters
epochs = 100
lr = 1e-3
batch_size = 64
img_dims = (96,96,3)

data = []
labels = []

# load image files from the dataset
image_files = [f for f in glob.glob(r'C:\Users\Cagdas\Desktop\gender\gender_dataset_face' + "/**/*", recursive=True) if not os.path.isdir(f)]
random.shuffle(image_files)

# converting images to arrays and labelling the categories
for img in image_files:

    image = cv2.imread(img)
    
    image = cv2.resize(image, (img_dims[0],img_dims[1]))
    image = img_to_array(image)
    data.append(image)

    label = img.split(os.path.sep)[-2] # C:\Files\gender_dataset_face\woman\face_1162.jpg
    if label == "woman":
        label = 1
    else:
        label = 0
        
    labels.append([label]) # [[1], [0], [0], ...]

# pre-processing
data = np.array(data, dtype="float") / 255.0
labels = np.array(labels)

# split dataset for training and validation
(trainX, testX, trainY, testY) = train_test_split(data, labels, test_size=0.2,
                                                  random_state=42)

trainY = to_categorical(trainY, num_classes=2) # [[1, 0], [0, 1], [0, 1], ...]
testY = to_categorical(testY, num_classes=2)

# Veri artırma işlemi 
aug = ImageDataGenerator(
    rotation_range=20,        
    width_shift_range=0.2,    
    height_shift_range=0.2,    
    shear_range=0.2,           
    zoom_range=0.2,            
    horizontal_flip=True,
    fill_mode="nearest"       
)

# define model
def build(width, height, depth, classes):
    model = Sequential()
    inputShape = (height, width, depth)
    chanDim = -1

    if K.image_data_format() == "channels_first": #Returns a string, either 'channels_first' or 'channels_last'
        inputShape = (depth, height, width)
        chanDim = 1
    
    # The axis that should be normalized, after a Conv2D layer with data_format="channels_first", 
    # set axis=1 in BatchNormalization.

    # Katmanların eklenmesi
    model.add(Input(shape=inputShape))                                # Giriş katmanı
    model.add(Conv2D(32, (3,3), padding="same"))                      # Konvolüsyon katmanı (32 filtre, 3x3 boyut)
    model.add(Activation("relu"))                                     # Aktivasyon fonksiyonu (ReLU)
    model.add(BatchNormalization(axis=chanDim))                       # Batch normalization
    model.add(MaxPooling2D(pool_size=(3,3)))                          # Max pooling (3x3 boyut)
    model.add(Dropout(0.25))                                          # Dropout (%25)

    model.add(Conv2D(64, (3,3), padding="same"))                      # 2. Konvolüsyon katmanı
    model.add(Activation("relu"))
    model.add(BatchNormalization(axis=chanDim))

    model.add(Conv2D(64, (3,3), padding="same"))                      # 3. Konvolüsyon katmanı
    model.add(Activation("relu"))
    model.add(BatchNormalization(axis=chanDim))
    model.add(MaxPooling2D(pool_size=(2,2)))                          # Max pooling (2x2 boyut)
    model.add(Dropout(0.25))                                          # Dropout (%25)

    model.add(Conv2D(128, (3,3), padding="same"))                     # 4. Konvolüsyon katmanı
    model.add(Activation("relu"))
    model.add(BatchNormalization(axis=chanDim))

    model.add(Conv2D(128, (3,3), padding="same"))                     # 5. Konvolüsyon katmanı
    model.add(Activation("relu"))
    model.add(BatchNormalization(axis=chanDim))
    model.add(MaxPooling2D(pool_size=(2,2)))                          # Max pooling (2x2 boyut)
    model.add(Dropout(0.25))                                          # Dropout (%25)

    model.add(Flatten())                                              # Verileri düzleştirme (1D vektöre dönüştürme)
    model.add(Dense(1024))                                            # Yoğun katman (1024 düğüm)
    model.add(Activation("relu"))
    model.add(BatchNormalization())
    model.add(Dropout(0.5))                                           # Dropout (%50)

    model.add(Dense(classes))                                         # Çıkış katmanı (sınıf sayısı kadar düğüm)
    model.add(Activation("sigmoid"))                                  # Sigmoid aktivasyon fonksiyonu

    return model

# build model
model = build(width=img_dims[0], height=img_dims[1], depth=img_dims[2],
                            classes=2)

# Modeli derleme
opt = Adam(learning_rate=lr)
model.compile(loss="binary_crossentropy", optimizer=opt, metrics=["accuracy"])

# steps_per_epoch'u doğru şekilde hesapla
steps_per_epoch = int(np.ceil(len(trainX) / batch_size))

class_weights = compute_class_weight('balanced', classes=np.unique(labels), y=labels.flatten())
class_weights_dict = dict(enumerate(class_weights))

print(f"Class Weights: {class_weights_dict}") # Class weights çıktısını kontrol etmek için ekleyin

# ReduceLROnPlateau Callback
reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',        # Doğrulama kaybına odaklan
    factor=0.5,                # Öğrenme oranını yarıya düşür
    patience=3,                # 3 epoch boyunca iyileşme olmazsa azalt
    verbose=1,                 # Azaltma sırasında bilgi ver
    min_lr=1e-6                # Minimum öğrenme oranı
)

# Modeli eğitme
H = model.fit(
    aug.flow(trainX, trainY, batch_size=batch_size),  # Veri artırma
    validation_data=(testX, testY),  # Doğrulama verileri
    steps_per_epoch=steps_per_epoch,  # Adım sayısı
    class_weight=class_weights_dict,
    epochs=epochs,  # Toplam epoch sayısı
    callbacks=[reduce_lr],                           # ReduceLROnPlateau callback
    verbose=1  # Eğitim ilerlemesi gösterilsin
)

```

```python
model.save('gender_detection.h5')
```

```python
sns.countplot(x=labels.flatten())
plt.title("Sınıf Dağılımı")
plt.show()

plt.style.use("ggplot")
plt.figure()
N = epochs
plt.plot(np.arange(0, N), H.history["loss"], label="train_loss")
plt.plot(np.arange(0, N), H.history["val_loss"], label="val_loss")
plt.plot(np.arange(0, N), H.history["accuracy"], label="train_accuracy")
plt.plot(np.arange(0, N), H.history["val_accuracy"], label="val_accuracy")

plt.title("Training Loss and Accuracy")
plt.xlabel("Epoch #")
plt.ylabel("Loss/Accuracy")
plt.legend(loc="upper right")

# save plot to disk
plt.savefig('plot.png')
plt.show()

# Eğitim sırasında modelin başarı oranlarını öğrenme
train_accuracy = H.history["accuracy"][-1]  # Son epoch'taki eğitim doğruluğu
val_accuracy = H.history["val_accuracy"][-1]  # Son epoch'taki doğrulama doğruluğu

print(f"Son Eğitim Başarı Oranı: {train_accuracy * 100:.2f}%")
print(f"Son Doğrulama Başarı Oranı: {val_accuracy * 100:.2f}%")
```

```python
import cv2, numpy as np, cvlib as cv
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import img_to_array
from PIL import Image, ImageTk
import tkinter as tk
import os

class GenderApp:
    def __init__(self, model_path):
        # Model yükle
        if not os.path.exists(model_path):
            raise FileNotFoundError(f"Model bulunamadı: {model_path}")
        self.model = load_model(model_path)
        self.classes = ['man','woman']

        # State
        self.running = False
        self.vs = None

        # Tkinter UI
        self.root = tk.Tk()
        self.root.title("Cinsiyet Tespiti")

        self.canvas = tk.Canvas(self.root, width=640, height=480)
        self.canvas.pack()

        btn_frame = tk.Frame(self.root)
        btn_frame.pack(fill="x", pady=5)

        self.btn_start = tk.Button(btn_frame, text="Başlat", command=self.start)
        self.btn_start.pack(side="left", padx=5)
        self.btn_stop = tk.Button(btn_frame, text="Durdur", state="disabled", command=self.stop)
        self.btn_stop.pack(side="left", padx=5)
        self.btn_close = tk.Button(btn_frame, text="Kapat", command=self.on_close)
        self.btn_close.pack(side="right", padx=5)

        self.root.protocol("WM_DELETE_WINDOW", self.on_close)

    def start(self):
        if self.running: return
        self.vs = cv2.VideoCapture(0)
        if not self.vs.isOpened():
            tk.messagebox.showerror("Hata", "Kamera açılamadı.")
            return
        self.running = True
        self.btn_start.config(state="disabled")
        self.btn_stop.config(state="normal")
        self._update_frame()

    def _update_frame(self):
        if not self.running:
            return

        ret, frame = self.vs.read()
        if not ret:
            self.stop()
            return

        # Yüz tespiti
        faces, _ = cv.detect_face(frame)
        for box in faces:
            (startX, startY, endX, endY) = [int(v) for v in box]
            face = frame[startY:endY, startX:endX]
            if face.shape[0] < 10 or face.shape[1] < 10:
                continue

            # Ön işleme + tahmin
            face = cv2.resize(face, (96,96))
            face = face.astype("float")/255.0
            face = img_to_array(face)
            face = np.expand_dims(face, axis=0)

            conf = self.model.predict(face)[0]
            idx = np.argmax(conf)
            label = f"{self.classes[idx]}: {conf[idx]*100:.2f}%"

            # Çerçeve + etiket
            y = startY - 10 if startY-10 > 10 else startY+10
            cv2.rectangle(frame, (startX,startY), (endX,endY), (0,255,0), 2)
            cv2.putText(frame, label, (startX, y),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,255,0), 2)

        # BGR→RGB → PIL → Tkinter
        rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        img = ImageTk.PhotoImage(Image.fromarray(rgb))
        self.canvas.create_image(0,0,anchor=tk.NW, image=img)
        self.canvas.image = img  # prevent GC

        self.root.after(15, self._update_frame)

    def stop(self):
        if not self.running: return
        self.running = False
        self.btn_start.config(state="normal")
        self.btn_stop.config(state="disabled")
        if self.vs:
            self.vs.release()

    def on_close(self):
        self.stop()
        self.root.destroy()

if __name__=="__main__":
    MODEL = "gender_detection.keras"
    app = GenderApp(MODEL)
    app.root.mainloop()
```
